```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from PIL import Image
import os
import glob
```

```python
path = "./dataset/"
jpeg_files = glob.glob(os.path.join(path, '*.jpeg'))
dataset = []
#Array de 114 elementos por cada imagen
for file in jpeg_files:
    image = Image.open(file).convert('L')
    array = np.array(image)
    array = array.flatten()
    #array = array.tolist()
    dataset.append(array)

dataset = np.array(dataset)
```

```python
n_spells = 3
np.random.seed(1)

kmeans = KMeans(n_clusters=n_spells,init='random')
kmeans.fit(dataset)
Z = kmeans.predict(dataset)

for i in range(0,n_spells):

    row = np.where(Z==i)[0]       # row in Z for elements of cluster i
    num = row.shape[0]            #  number of elements for each cluster
    r = int(np.floor(num/10.))    # number of rows in the figure of the cluster 

    print("cluster "+str(i))
    print(str(num)+" elements")

    plt.figure(figsize=(10,10))
    for k in range(0, num):
        plt.subplot(r+1, 10, k+1)
        image = dataset[row[k], ]
        image = image.reshape(19, 6)
        plt.imshow(image, cmap='gray')
        plt.axis('off')
    plt.show()
```
